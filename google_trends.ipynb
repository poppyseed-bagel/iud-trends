This notebook displays the code that was used to extract Google trends data from the Google Trends API.

```python
# Import necessary packages
import pandas as pd                        
from pytrends.request import TrendReq
```

```python
# Instantiate the class
pytrend = TrendReq()
```

```python
# keywords we are interested in
kw_list=['IUD', 'birth control', 'kyleena', 'mirena', 'IUS']

# object that looks at the keywords in Great Britain overtime
pytrend.build_payload(kw_list, geo = 'GB-ENG')

# search interest for overtime in the above object
kw_df = pytrend.interest_over_time()
kw_df
```

```python
kw_df.plot()
```
